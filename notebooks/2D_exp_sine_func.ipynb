```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from kan import KAN, create_dataset
import time

# FIX 1: Force CPU. CUDA interpolation during refine() often causes NaNs in pykan
device = torch.device('cpu')
torch.set_default_dtype(torch.float64)

# 1. Define the synthetic dataset and force it to CPU
f = lambda x: torch.exp(torch.sin(torch.pi * x[:, [0]]) + x[:, [1]]**2)
dataset = create_dataset(f, n_var=2, train_num=1000, test_num=1000, device=device)

# 2. Experiment Configuration
grids = [3, 5, 10, 20, 50, 100, 200, 500, 1000]
steps_per_grid = 200
adam_steps = 50
lbfgs_steps = steps_per_grid - adam_steps

def train_kan_with_grid_extension(arch, grids):
    # FIX 2: Explicitly initialize the model on the CPU
    model = KAN(width=arch, grid=grids[0], k=3, seed=42, device=device)

    results = {
        'train_rmse': [],
        'test_rmse': [],
        'times': [],
        'grid_test_loss': []
    }

    for g in grids:
        print(f"Training architecture {arch} on grid size {g}...")
        if g != grids[0]:
            # Fine-grain the spline grids
            model = model.refine(g)

        start_time = time.time()

        # FIX 3: Lower Adam learning rate slightly to ensure gentle warm-up
        res_adam = model.fit(dataset, opt="Adam", steps=adam_steps, update_grid=False, lr=0.001)
        res_lbfgs = model.fit(dataset, opt="LBFGS", steps=lbfgs_steps, update_grid=False)

        elapsed = time.time() - start_time

        # Combine the loss histories from both optimizers
        train_loss = np.concatenate((res_adam['train_loss'], res_lbfgs['train_loss']))
        test_loss = np.concatenate((res_adam['test_loss'], res_lbfgs['test_loss']))

        # PyKAN natively returns MSE; we take sqrt for RMSE
        train_rmse = np.sqrt(train_loss)
        test_rmse = np.sqrt(test_loss)

        results['train_rmse'].extend(train_rmse)
        results['test_rmse'].extend(test_rmse)
        results['times'].append(elapsed)
        results['grid_test_loss'].append(test_rmse[-1])

    return results

# 3. Run Experiments
results_251 = train_kan_with_grid_extension([2, 5, 1], grids)
results_211 = train_kan_with_grid_extension([2, 1, 1], grids)

# 4. Plotting
fig, axs = plt.subplots(2, 2, figsize=(14, 9))
fig.suptitle(r'Fitting $f(x, y) = \exp(\sin(\pi x) + y^2)$', fontsize=16)

steps_total = np.arange(len(grids) * steps_per_grid)

# Helper function for top plots
def plot_training_dynamics(ax, results, title, threshold_step):
    ax.semilogy(steps_total, results['train_rmse'], label='train', color='steelblue')
    ax.semilogy(steps_total, results['test_rmse'], label='test', color='sandybrown')

    # Add grid text and vertical threshold lines
    for i, g in enumerate(grids):
        ax.text(i * steps_per_grid + 50, 1e-8, f'grid={g}', rotation=60, fontsize=9)

    ax.axvline(x=threshold_step, color='red', linestyle='--', alpha=0.6)
    ax.text(threshold_step - 40, 1e-6, 'interpolation threshold', color='red', rotation=90, alpha=0.6)

    ax.set_title(title, fontsize=14)
    ax.set_xlabel('step', fontsize=12)
    ax.set_ylabel('RMSE', fontsize=12)
    ax.set_ylim(1e-9, 1e1)
    ax.legend()

# Top Left: [2, 5, 1] Dynamics
plot_training_dynamics(axs[0, 0], results_251, 'KAN [2,5,1]', threshold_step=900)

# Top Right: [2, 1, 1] Dynamics
plot_training_dynamics(axs[0, 1], results_211, 'KAN [2,1,1]', threshold_step=1100)

# Bottom Left: Test Loss vs Grid Size (Scaling Laws)
G = np.array(grids)
axs[1, 0].loglog(G, results_251['grid_test_loss'], 'k-o', label='KAN [2,5,1] sqrt(mean of squared)', markersize=5)
axs[1, 0].loglog(G, results_211['grid_test_loss'], 'b-o', label='KAN [2,1,1] sqrt(mean of squared)', markersize=5)

# Add reference lines for scaling laws
axs[1, 0].loglog(G[2:7], 1e-1 * G[2:7]**(-2.0), 'k--', alpha=0.8)
axs[1, 0].text(7, 2e-4, '$G^{-2}$')

axs[1, 0].loglog(G[3:8], 1e0 * G[3:8]**(-3.0), 'b--', alpha=0.8)
axs[1, 0].text(15, 5e-4, '$G^{-3}$', color='blue')

axs[1, 0].loglog(G[3:8], 2e-1 * G[3:8]**(-4.0), 'c--', alpha=0.5)
axs[1, 0].text(20, 8e-7, '$G^{-4}$', color='c')

axs[1, 0].set_xlabel('grid size $G$', fontsize=12)
axs[1, 0].set_ylabel('test loss', fontsize=12)
axs[1, 0].legend()

# Bottom Right: Training Time vs Grid Size
axs[1, 1].loglog(G, np.array(results_251['times']) / steps_per_grid, 'k-o', label='KAN [2,5,1]', markersize=5)
axs[1, 1].loglog(G, np.array(results_211['times']) / steps_per_grid, 'b-o', label='KAN [2,1,1]', markersize=5)
axs[1, 1].set_xlabel('grid size $G$', fontsize=12)
axs[1, 1].set_ylabel('training time (seconds/step)', fontsize=12)
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from kan import KAN, create_dataset
import time

# Force CPU and float64 for mathematical stability during spline interpolation
device = torch.device('cpu')
torch.set_default_dtype(torch.float64)

# 1. Define the synthetic dataset
# f(x,y) = exp(sin(pi * x) + y^2)
f = lambda x: torch.exp(torch.sin(torch.pi * x[:, [0]]) + x[:, [1]]**2)
dataset = create_dataset(f, n_var=2, train_num=1000, test_num=1000, device=device)

# 2. Experiment Configuration
grids = [3, 5, 10, 20, 50, 100, 200, 500, 1000]
steps_per_grid = 200

def train_kan_with_grid_extension(arch, grids, seed=0):
    # CRITICAL: Use the author's seed=0.
    # With a narrow bottleneck like [2,1,1], a bad seed causes LBFGS to flatline.
    model = KAN(width=arch, grid=grids[0], k=3, seed=seed, device=device)

    results = {
        'train_rmse': [],
        'test_rmse': [],
        'times': [],
        'grid_test_loss': []
    }

    for g in grids:
        print(f"Training architecture {arch} on grid size {g}...")
        if g != grids[0]:
            # Fine-grain the spline grids
            model = model.refine(g)

        start_time = time.time()

        # FIX 1: Removed update_grid=False so splines can adapt to shifting activations
        # FIX 2: lamb=0.0 ensures pure MSE loss without sparsity penalties
        res = model.fit(dataset, opt="LBFGS", steps=steps_per_grid, lamb=0.0)

        elapsed = time.time() - start_time

        # PyKAN natively returns MSE; we take sqrt for RMSE
        train_rmse = np.sqrt(res['train_loss'])
        test_rmse = np.sqrt(res['test_loss'])

        results['train_rmse'].extend(train_rmse)
        results['test_rmse'].extend(test_rmse)
        results['times'].append(elapsed)
        results['grid_test_loss'].append(test_rmse[-1])

    return results

# 3. Run Experiments
results_251 = train_kan_with_grid_extension([2, 5, 1], grids, seed=0)
results_211 = train_kan_with_grid_extension([2, 1, 1], grids, seed=0)

# 4. Plotting
fig, axs = plt.subplots(2, 2, figsize=(14, 9))
fig.suptitle(r'Fitting $f(x, y) = \exp(\sin(\pi x) + y^2)$', fontsize=16)

steps_total = np.arange(len(grids) * steps_per_grid)

# Helper function for top plots
def plot_training_dynamics(ax, results, title, threshold_step):
    ax.semilogy(steps_total, results['train_rmse'], label='train', color='steelblue')
    ax.semilogy(steps_total, results['test_rmse'], label='test', color='sandybrown')

    # Add grid text and vertical threshold lines
    for i, g in enumerate(grids):
        ax.text(i * steps_per_grid + 50, 1e-8, f'grid={g}', rotation=60, fontsize=9)

    ax.axvline(x=threshold_step, color='red', linestyle='--', alpha=0.6)
    ax.text(threshold_step - 40, 1e-6, 'interpolation threshold', color='red', rotation=90, alpha=0.6)

    ax.set_title(title, fontsize=14)
    ax.set_xlabel('step', fontsize=12)
    ax.set_ylabel('RMSE', fontsize=12)
    ax.set_ylim(1e-9, 1e1)
    ax.legend()

# Top Left: [2, 5, 1] Dynamics
plot_training_dynamics(axs[0, 0], results_251, 'KAN [2,5,1]', threshold_step=900)

# Top Right: [2, 1, 1] Dynamics
plot_training_dynamics(axs[0, 1], results_211, 'KAN [2,1,1]', threshold_step=1100)

# Bottom Left: Test Loss vs Grid Size (Scaling Laws)
G = np.array(grids)
axs[1, 0].loglog(G, results_251['grid_test_loss'], 'k-o', label='KAN [2,5,1] sqrt(mean of squared)', markersize=5)
axs[1, 0].loglog(G, results_211['grid_test_loss'], 'b-o', label='KAN [2,1,1] sqrt(mean of squared)', markersize=5)

# Add reference lines for scaling laws
axs[1, 0].loglog(G[2:7], 1e-1 * G[2:7]**(-2.0), 'k--', alpha=0.8)
axs[1, 0].text(7, 2e-4, '$G^{-2}$')

axs[1, 0].loglog(G[3:8], 1e0 * G[3:8]**(-3.0), 'b--', alpha=0.8)
axs[1, 0].text(15, 5e-4, '$G^{-3}$', color='blue')

axs[1, 0].loglog(G[3:8], 2e-1 * G[3:8]**(-4.0), 'c--', alpha=0.5)
axs[1, 0].text(20, 8e-7, '$G^{-4}$', color='c')

axs[1, 0].set_xlabel('grid size $G$', fontsize=12)
axs[1, 0].set_ylabel('test loss', fontsize=12)
axs[1, 0].legend()

# Bottom Right: Training Time vs Grid Size
axs[1, 1].loglog(G, np.array(results_251['times']) / steps_per_grid, 'k-o', label='KAN [2,5,1]', markersize=5)
axs[1, 1].loglog(G, np.array(results_211['times']) / steps_per_grid, 'b-o', label='KAN [2,1,1]', markersize=5)
axs[1, 1].set_xlabel('grid size $G$', fontsize=12)
axs[1, 1].set_ylabel('training time (seconds/step)', fontsize=12)
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
# Installation (if needed):
#   pip install pykan matplotlib torch

import torch
import numpy as np
import time
import matplotlib.pyplot as plt
from kan import KAN
from kan.utils import create_dataset

# Set random seeds for reproducibility
seed = 0
torch.manual_seed(seed)
np.random.seed(seed)
# Force CPU usage
device = torch.device("cpu")

# Define target function f(x,y)
f = lambda x: torch.exp(torch.sin(torch.pi * x[:, [0]]) + x[:, [1]]**2)

# Create dataset: 1000 train, 1000 test samples in [-1,1]^2
dataset = create_dataset(f, n_var=2, train_num=1000, test_num=1000, device=device, seed=seed)

# Grid sizes to use (number of intervals)
grid_sizes = [3, 5, 10, 20, 50, 100, 200, 500, 1000]

# Lists to record metrics
all_train_rmse = []
all_test_rmse = []
time_per_step = []

model = None
step_offset = 0
for idx, G in enumerate(grid_sizes):
    if idx == 0:
        # Initialize first model with coarse grid
        model = KAN(width=[2,5,1], grid=G, k=3, device=device, seed=seed)
    else:
        # Create new model with finer grid and initialize from previous
        new_model = KAN(width=[2,5,1], grid=G, k=3, device=device, seed=seed)
        new_model.initialize_from_another_model(model, dataset['train_input'])
        model = new_model

    # Train for 200 LBFGS steps
    start_time = time.perf_counter()
    results = model.fit(dataset, opt='LBFGS', steps=200, update_grid=False, lamb=0.0)
    elapsed = time.perf_counter() - start_time

    # Append RMSE history
    train_loss = np.array(results['train_loss'])
    test_loss  = np.array(results['test_loss'])
    all_train_rmse.extend(train_loss.tolist())
    all_test_rmse.extend(test_loss.tolist())
    # Record time per step (seconds)
    time_per_step.append(elapsed / 200)
    print(f"Grid {G}: final train RMSE={train_loss[-1]:.4f}, test RMSE={test_loss[-1]:.4f}, time/step={elapsed/200:.4f}s")

# Compute final test RMSE for each grid (after 200 steps at that grid)
final_test_rmse = [all_test_rmse[(i+1)*200 - 1] for i in range(len(grid_sizes))]

# Plot 1: Training/Test RMSE vs Training Steps, marking grid changes
plt.figure(figsize=(6,4))
steps = np.arange(1, len(all_train_rmse)+1)
plt.plot(steps, all_train_rmse, label='Train RMSE')
plt.plot(steps, all_test_rmse, label='Test RMSE')
# Mark grid change points
for i in range(1, len(grid_sizes)):
    plt.axvline(i*200, color='gray', linestyle='--', alpha=0.7)
plt.xlabel('Training Step')
plt.ylabel('RMSE')
plt.title('Train/Test RMSE vs Steps')
plt.legend()
plt.tight_layout()

# Plot 2: Final test RMSE vs grid size (log–log) with reference slopes
plt.figure(figsize=(6,4))
Gs = np.array(grid_sizes, dtype=float)
plt.loglog(Gs, final_test_rmse, 'o-', label='Test RMSE')
# Reference lines: G^{-2}, G^{-3}, G^{-4} (arbitrary scale for illustration)
ref = Gs
plt.loglog(ref, ref**(-2)* (final_test_rmse[0]*ref[0]**2), '--', label=r'$G^{-2}$')
plt.loglog(ref, ref**(-3)* (final_test_rmse[0]*ref[0]**3), '--', label=r'$G^{-3}$')
plt.loglog(ref, ref**(-4)* (final_test_rmse[0]*ref[0]**4), '--', label=r'$G^{-4}$')
plt.xlabel('Grid size $G$ (intervals)')
plt.ylabel('Final Test RMSE')
plt.title('Final Test RMSE vs Grid Size (log–log)')
plt.legend()
plt.tight_layout()

# Plot 3: Time per step vs grid size (log–log)
plt.figure(figsize=(6,4))
plt.loglog(Gs, time_per_step, 'o-')
plt.xlabel('Grid size $G$ (intervals)')
plt.ylabel('Time per LBFGS step (s)')
plt.title('Time per Step vs Grid Size')
plt.tight_layout()

plt.show()
```
